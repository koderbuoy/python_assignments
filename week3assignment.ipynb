```python
#Question 1
balance = int(input("Balance: "))
daily_withdrawn = int(input("Already withdrawn today: "))
amount = int(input("Amount to withdraw: "))

if amount % 500 != 0:
    print("Invalid amount. Must be a multiple of NPR 500.")
elif amount > balance:
    print("Insufficient balance.")
elif daily_withdrawn + amount > 50000:
    print("Daily withdrawal limit reached.")
else:
    balance = balance - amount
    print("Withdrawal successful.")
    print("Your current balance after withdrawal: NPR", balance)
```

```python
#Question 2

total = float(input("Total purchase (NPR): "))
member = input("Loyalty member? (yes/no): ")

if total < 1000:
    discount = 0
elif total < 5000:
    discount = 5
elif total < 15000:
    discount = 10
else:
    discount = 20

price = total - (total * discount / 100)

if member == "yes":
    price = price - (price * 5 / 100)

print("Final payable amount: NPR", price)
```

```python
#Question 3

inventory = [
    {"item": "Rice", "stock": 5, "threshold": 10},
    {"item": "Eggs", "stock": 24, "threshold": 12},
    {"item": "Milk", "stock": 3, "threshold": 6},
    {"item": "Bread", "stock": 8, "threshold": 5},
    {"item": "Chicken", "stock": 0, "threshold": 4},
    {"item": "Cooking Oil", "stock": 2, "threshold": 3},
]

count = 0
for i in inventory:
    if i["stock"] < i["threshold"]:
        print("Restock alert:", i["item"])
        count = count + 1

print("Total items needing restock:", count)
```

```python
#Question 4

passwords = ["hello", "Hello123", "H3ll0@World", "12345678", "MyP@ss!"]

for p in passwords:
    problems = []
    if len(p) < 8:
        problems.append("at least 8 characters")
    if not any(c.isupper() for c in p):
        problems.append("one uppercase letter")
    if not any(c.islower() for c in p):
        problems.append("one lowercase letter")
    if not any(c.isdigit() for c in p):
        problems.append("one digit")
    if not any(c in "!@#$%^&*" for c in p):
        problems.append("one special character (!@#$%^&*)")

    if problems:
        print(p, "-> Weak. Missing:", ", ".join(problems))
    else:
        print(p, "-> Strong")
```

```python
#Question 5

trips = [
    {"distance": 1.5, "hour": 14},
    {"distance": 5.0, "hour": 22},
    {"distance": 12.0, "hour": 3},
    {"distance": 8.5, "hour": 10},
    {"distance": 2.0, "hour": 23},
]

for t in trips:
    d = t["distance"]
    h = t["hour"]

    fare = 150
    if d > 2:
        extra = d - 2
        if extra <= 8:
            fare = fare + extra * 35
        else:
            fare = fare + 8 * 35 + (extra - 8) * 28

    if h >= 22 or h < 5:
        fare = fare + fare * 10 / 100

    print("Distance:", d, "km | Hour:", h, "| Fare: NPR", fare)
```
