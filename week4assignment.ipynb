```python
#Question 1

inventory = {
    "rice":  {"price": 120, "stock": 20},
    "milk":  {"price": 90,  "stock": 10},
    "bread": {"price": 60,  "stock": 15},
    "eggs":  {"price": 15,  "stock": 30}
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}

def process_order(inventory, cart):
    total = 0
    print("---- Bill ----")
    for item in cart:
        qty = cart[item]
        if inventory[item]["stock"] < qty:
            print("Sorry, not enough stock for", item)
        else:
            cost = inventory[item]["price"] * qty
            total = total + cost
            inventory[item]["stock"] = inventory[item]["stock"] - qty
            print(item, "x" + str(qty), "= NPR", cost)
    print("Grand Total: NPR", total)
    print("--------------")
    stock_info = ", ".join(i + "=" + str(inventory[i]["stock"]) for i in inventory)
    print("Updated stock:", stock_info)

process_order(inventory, cart)
```

```python
#Question 2
sensors = [
    ("Chatara", 2.8),
    ("Tribeni Ghat", 5.4),
    ("Koshi Barrage", 4.1),
    ("Sunsari Bridge", 1.9),
    ("Saptakoshi Camp", 6.0),
]

def check_water_level(location, level):
    if level < 3:
        return "Safe"
    elif level <= 5:
        return "Warning — Alert nearby villages"
    else:
        return "DANGER — Evacuate immediately!"

for loc, lvl in sensors:
    print(loc, "(" + str(lvl) + " m):", check_water_level(loc, lvl))
```

```python
#Question 3

bs_months = ["Baisakh","Jestha","Ashadh","Shrawan","Bhadra","Ashwin",
             "Kartik","Mangsir","Poush","Magh","Falgun","Chaitra"]

customers = [
    {"name": "Ramesh Thapa",  "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki",  "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai",    "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"},
]

def convert_date(date_str, from_cal, to_cal):
    parts = date_str.split("-")
    y, m, d = int(parts[0]), int(parts[1]), int(parts[2])
    if from_cal == to_cal:
        return date_str
    if from_cal == "AD":
        y = y + 56
    else:
        y = y - 56
    return str(y) + "-" + str(m).zfill(2) + "-" + str(d).zfill(2)

for c in customers:
    result = convert_date(c["date"], c["cal"], c["need"])
    parts = result.split("-")
    y, m, d = int(parts[0]), int(parts[1]), int(parts[2])

    if c["style"] == "iso":
        formatted = result + " " + c["need"]
    elif c["style"] == "full":
        formatted = str(d) + "th " + bs_months[m - 1] + ", " + str(y) + " " + c["need"]
    else:
        formatted = str(y) + " " + bs_months[m - 1] + " " + str(d) + " " + c["need"]

    print(c["name"], "| Original:", c["date"], c["cal"], "| Converted:", formatted)
```
