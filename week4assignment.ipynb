```python
#Question 1

inventory = {
    "rice":  {"price": 120, "stock": 20},
    "milk":  {"price": 90,  "stock": 10},
    "bread": {"price": 60,  "stock": 15},
    "eggs":  {"price": 15,  "stock": 30}
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}

def process_order(inventory, cart):
    total = 0
    print("---- Bill ----")
    for item in cart:
        qty = cart[item]
        if inventory[item]["stock"] < qty:
            print("Sorry, not enough stock for", item)
        else:
            cost = inventory[item]["price"] * qty
            total = total + cost
            inventory[item]["stock"] = inventory[item]["stock"] - qty
            print(item, "x" + str(qty), "= NPR", cost)
    print("Grand Total: NPR", total)
    print("--------------")
    stock_info = ", ".join(i + "=" + str(inventory[i]["stock"]) for i in inventory)
    print("Updated stock:", stock_info)

process_order(inventory, cart)
```

```python
#Question 3

bs_months = ["Baisakh","Jestha","Ashadh","Shrawan","Bhadra","Ashwin",
             "Kartik","Mangsir","Poush","Magh","Falgun","Chaitra"]

customers = [
    {"name": "Ramesh Thapa",  "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki",  "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai",    "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"},
]

def convert_date(date_str, from_cal, to_cal):
    parts = date_str.split("-")
    y, m, d = int(parts[0]), int(parts[1]), int(parts[2])
    if from_cal == to_cal:
        return date_str
    if from_cal == "AD":
        y = y + 56
    else:
        y = y - 56
    return str(y) + "-" + str(m).zfill(2) + "-" + str(d).zfill(2)

for c in customers:
    result = convert_date(c["date"], c["cal"], c["need"])
    parts = result.split("-")
    y, m, d = int(parts[0]), int(parts[1]), int(parts[2])

    if c["style"] == "iso":
        formatted = result + " " + c["need"]
    elif c["style"] == "full":
        formatted = str(d) + "th " + bs_months[m - 1] + ", " + str(y) + " " + c["need"]
    else:
        formatted = str(y) + " " + bs_months[m - 1] + " " + str(d) + " " + c["need"]

    print(c["name"], "| Original:", c["date"], c["cal"], "| Converted:", formatted)
```

```python
#Question 4
text = """
Nepal is a beautiful country. Nepal has Mount Everest.
Everest is the highest mountain in the world. Many tourists
visit Nepal every year to see Everest and other mountains.
Nepal is known for its mountains and natural beauty.
"""

def word_frequency(text):
    words = text.lower().replace(".", "").replace(",", "").split()
    counts = {}
    for w in words:
        if w in counts:
            counts[w] = counts[w] + 1
        else:
            counts[w] = 1
    top3 = sorted(counts, key=lambda x: counts[x], reverse=True)[:3]
    print("Top 3 words:")
    for w in top3:
        print(w, "—", counts[w], "times")

word_frequency(text)
```

```python
#Question 5
accounts = {
    "A001": {"name": "Ramesh Thapa",  "balance": 15000, "pin": "1234"},
    "A002": {"name": "Sunita Karki",  "balance": 8500,  "pin": "5678"},
    "A003": {"name": "Bikash Rai",    "balance": 22000, "pin": "9012"}
}

def atm(account_id, pin, action, amount=0):
    if account_id not in accounts:
        print("Account not found")
        return
    acc = accounts[account_id]
    if acc["pin"] != pin:
        print("Incorrect PIN")
        return
    if action == "balance":
        print(acc["name"], "| Balance: NPR", acc["balance"])
    elif action == "deposit":
        acc["balance"] = acc["balance"] + amount
        print("Deposited NPR", amount, "| New balance: NPR", acc["balance"])
    elif action == "withdraw":
        if amount > acc["balance"]:
            print("Insufficient funds")
        else:
            acc["balance"] = acc["balance"] - amount
            print("Withdrawn NPR", amount, "| New balance: NPR", acc["balance"])

atm("A001", "1234", "balance")
atm("A002", "0000", "withdraw", 2000)
atm("A002", "5678", "deposit",  3000)
atm("A003", "9012", "withdraw", 25000)
atm("A004", "1111", "balance")
```
